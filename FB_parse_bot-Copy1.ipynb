```python
%%time

results = []

for i, post_id in enumerate(post_ids):
    row = []
    try:
        browser.get("http://facebook.com/" + post_id)
        sleep(1)
        row += [post_id]
        row += [browser.find_element_by_xpath('/html/body/div[1]/div/div[1]/div/div[3]/div/div/div[1]/div[1]/div[4]/div[1]/div/div/div/div/div/div/div/div/div/div[1]/div/div[2]/div/div[4]/div[1]').text]
        row += [browser.find_element_by_xpath('/html/body/div[1]/div/div[1]/div/div[3]/div/div/div[1]/div[1]/div[4]/div[1]/div/div/div/div/div/div/div/div/div/div[1]/div/div[2]/div/div[4]/div[2]/div/div/div[1]').text]
        try:
            row += [{i.split(': ')[0]:int(i.split(': ')[1]) for i in ['likes: '+x if i==0 else x for i, x in enumerate(browser.find_element_by_xpath('/html/body/div[1]/div/div[1]/div/div[3]/div/div/div[1]/div[1]/div[4]/div[1]/div/div/div/div/div/div/div/div/div/div[1]/div/div[2]/div/div[5]/div/div/div[1]/div/div[1]/div').text.split('\n')[1:])]}]
        except:
            row += [{'likes': 0, 'Комментарии': 0, 'Поделились': 0}]
    except:
        row = [post_id, '0\nОхват (люди)', '0', {}]
    
    results.append(row)
    print(i, len(post_ids), row)
```

```python
import datetime

df = pd.DataFrame(results, columns=['post_id', 'coverage', 'interactions', 'lcs', ])
df = df.set_index('post_id')
df.to_excel('excel/FULL_facebook_dates_likes_comments.xlsx')
df
```
