```python
%%time

results = []

for i, post_id in enumerate(post_ids):
    row = []
    try:
        browser.get("http://facebook.com/" + post_id)
        sleep(1)
        row += [post_id]
        row += [browser.find_element_by_xpath('/html/body/div[1]/div/div[1]/div/div[3]/div/div/div[1]/div[1]/div[4]/div[1]/div/div/div/div/div/div/div/div/div/div[1]/div/div[2]/div/div[4]/div[1]').text]
        row += [browser.find_element_by_xpath('/html/body/div[1]/div/div[1]/div/div[3]/div/div/div[1]/div[1]/div[4]/div[1]/div/div/div/div/div/div/div/div/div/div[1]/div/div[2]/div/div[4]/div[2]/div/div/div[1]').text]
        try:
            row += [{i.split(': ')[0]:int(i.split(': ')[1]) for i in ['likes: '+x if i==0 else x for i, x in enumerate(browser.find_element_by_xpath('/html/body/div[1]/div/div[1]/div/div[3]/div/div/div[1]/div[1]/div[4]/div[1]/div/div/div/div/div/div/div/div/div/div[1]/div/div[2]/div/div[5]/div/div/div[1]/div/div[1]/div').text.split('\n')[1:])]}]
        except:
            row += [{'likes': 0, 'Комментарии': 0, 'Поделились': 0}]
    except:
        row = [post_id, '0\nОхват (люди)', '0', {}]
    
    results.append(row)
    print(i, len(post_ids), row)
```

```python
import datetime

df = pd.DataFrame(results, columns=['post_id', 'coverage', 'interactions', 'lcs', ])
df = df.set_index('post_id')
df.to_excel('excel/FULL_facebook_dates_likes_comments.xlsx')
df
```

```python
last_df = pd.read_csv('dataset/facebook_yessenov_2.csv').set_index('id')
last_df
```

```python
res_df = pd.concat([last_df, df], axis=1)
res_df
```

```python
res_df = res_df[res_df.interactions != '0'][['created_time', 'message', 'coverage', 'interactions', 'lcs']]
res_df
```

```python
from datetime import datetime

res_df['datetime'] = res_df.created_time.apply(lambda x: datetime.strptime(x[:-5], '%Y-%m-%dT%H:%M:%S').strftime('%d.%m.%Y'))
res_df['hour_minute_seconds'] = res_df.created_time.apply(lambda x: datetime.strptime(x[:-5], '%Y-%m-%dT%H:%M:%S').strftime('%H:%M:%S'))
res_df
```

```python
lcs = pd.DataFrame(res_df.lcs.values.tolist())
lcs.index = res_df.index
lcs = lcs.fillna(0)
lcs
```

```python
pd.concat([res_df, lcs], axis=1).columns
```

```python
saving_df = pd.concat([res_df, lcs], axis=1)
savinf_df = saving_df[['created_time', 'message', 'datetime', 'hour_minute_seconds', 'likes', 
                       'Комментарии', 'Поделились', 'coverage', 'interactions', ]]
savinf_df
```

```python
savinf_df['coverage'] = savinf_df.coverage.apply(lambda x: x.split('\n')[0])
savinf_df['message'] = savinf_df.message.fillna('no data')
savinf_df
```

```python

savinf_df.to_excel('excel/FULL_cleaned_facebook_dates_likes_comments.xlsx')
```
